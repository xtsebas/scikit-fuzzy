## fuzz.trimf

La función `fuzz.trimf` define una función de membresía triangular. Es útil cuando se desea representar un conjunto difuso en forma de triángulo, con un valor mínimo, un valor medio y un valor máximo.

### Parámetros

- **x**: array-like
  - El conjunto de valores sobre los cuales se evaluará la función de membresía.
- **a**: float
  - El valor mínimo del triángulo.
- **b**: float
  - El valor medio del triángulo.
- **c**: float
  - El valor máximo del triángulo.

### Retorno

- **y**: array-like
  - Los grados de pertenencia correspondientes a cada valor en `x`.

### Ejemplo de uso


```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

x = np.arange(0, 11, 1)

triangular = fuzz.trimf(x, [0, 5, 10])

plt.plot(x, triangular, label='Triangular')
plt.xlabel('x')
plt.ylabel('Membresía')
plt.title('Función de Membresía Triangular')
plt.legend()
plt.show()
```

## fuzz.interp_membership

La función `fuzz.interp_membership` realiza la difusificación de un valor en una función de membresía. Nos dice qué tan bien pertenece un valor específico a un conjunto difuso.

### Parámetros

- **x**: array-like
  - El conjunto de valores sobre los cuales se evaluará la función de membresía.
- **mfx**: array-like
  - Los grados de pertenencia correspondientes a cada valor en `x`.
- **xq**: float
  - El valor específico que se desea evaluar.

### Retorno

- **y**: float
  - El grado de pertenencia del valor `xq` en la función de membresía `mfx`.

### Ejemplo de uso


```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

nivel_membresia = fuzz.interp_membership(x, triangular, 7)
print("Nivel de membresía de 7:", nivel_membresia)
```

## fuzz.defuzz

La función `fuzz.defuzz` es utilizada para defuzzificar, es decir, para convertir un conjunto difuso en un valor crisp. Hay varios métodos de defuzzificación, como el método del centroide.

### Parámetros

- **x**: array-like
  - El conjunto de valores sobre los cuales se evaluará la función de membresía.
- **mfx**: array-like
  - Los grados de pertenencia correspondientes a cada valor en `x`.
- **mode**: string
  - El método de defuzzificación a utilizar. Algunos métodos comunes son 'centroid', 'bisector', 'mom', 'som', 'lom'.

### Retorno

- **y**: float
  - El valor crisp resultante de la defuzzificación.

### Ejemplo de uso


```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

x_propina = np.arange(0, 26, 1)
propina_baja = fuzz.trimf(x_propina, [0, 0, 13])
propina_media = fuzz.trimf(x_propina, [0, 13, 25])
activacion = np.fmax(propina_baja, propina_media)
propina_final = fuzz.defuzz(x_propina, activacion, 'centroid')
print("Propina final (centroid):", propina_final)
```

## fuzz.gaussmf

La función `fuzz.gaussmf` define una función de membresía gaussiana. 
Es útil cuando necesitas una función de membresía suave y simétrica en forma de campana.

### Parámetros

- **x**: array-like
  - El conjunto de valores sobre los cuales se evaluará la función de membresía.
- **mean**: float
  - La media de la distribución gaussiana.
- **sigma**: float
  - La desviación estándar de la distribución gaussiana.

### Retorno

- **y**: array-like
  - Los grados de pertenencia correspondientes a cada valor en `x`.

### Ejemplo de uso


```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.1)
gaussiana = fuzz.gaussmf(x, mean=5, sigma=1)

plt.plot(x, gaussiana, label='Gaussiana')
plt.xlabel('x')
plt.ylabel('Membresía')
plt.title('Función de Membresía Gaussiana')
plt.legend()
plt.show()
```

## fuzz.zmf y fuzz.smf

Estas funciones definen funciones de membresía en forma de “Z” y “S”, respectivamente. Son útiles para representar transiciones suaves y graduales en sistemas difusos.

### fuzz.zmf

La función `fuzz.zmf` es útil para representar una transición que cae de manera rápida, similar a la forma de una "Z".

#### Parámetros

- **x**: array-like
  - El conjunto de valores sobre los cuales se evaluará la función de membresía.
- **a**: float
  - El punto donde la función comienza a caer.
- **b**: float
  - El punto donde la función termina de caer.

#### Retorno

- **y**: array-like
  - Los grados de pertenencia correspondientes a cada valor en `x`.

#### Ejemplo de uso


```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.1)
forma_z = fuzz.zmf(x, 3, 7)

plt.plot(x, forma_z, label='Forma Z')
plt.xlabel('x')
plt.ylabel('Membresía')
plt.title('Función de Membresía en Forma de Z')
plt.legend()
plt.show()
```

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.1)
forma_s = fuzz.smf(x, 3, 7)

plt.plot(x, forma_s, label='Forma S')
plt.xlabel('x')
plt.ylabel('Membresía')
plt.title('Función de Membresía en Forma de S')
plt.legend()
plt.show()
```

## fuzz.dsigmf

La función `fuzz.dsigmf` crea una función de membresía sigmoide de diferencia doble. 
Es útil para casos en los que se necesita una función de membresía con dos transiciones sigmoides, una ascendente y otra descendente.

### Parámetros

- **x**: array-like
  - El conjunto de valores sobre los cuales se evaluará la función de membresía.
- **b1**: float
  - El punto de inflexión de la primera sigmoide.
- **c1**: float
  - La pendiente de la primera sigmoide.
- **b2**: float
  - El punto de inflexión de la segunda sigmoide.
- **c2**: float
  - La pendiente de la segunda sigmoide.

### Retorno

- **y**: array-like
  - Los grados de pertenencia correspondientes a cada valor en `x`.

### Ejemplo de uso


```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

x = np.arange(0, 10, 0.1)
dsigmoide = fuzz.dsigmf(x, -2, 4, 2, 6)

plt.plot(x, dsigmoide, label='Sigmoide Doble')
plt.xlabel('x')
plt.ylabel('Membresía')
plt.title('Función de Membresía Sigmoide Doble')
plt.legend()
plt.show()
```
